```python
def parse_ingredients(s): #needed because the special characters break ast function
    if not isinstance(s, str):
        return []
    # remove starting/ending brackets
    s = s.strip("[]")
    # split on comma
    items = s.split(",")
    # strip whitespace and quotes
    items = [item.strip().strip('"').strip("'") for item in items]
    # remove empty strings
    return [item for item in items if item]
super_df = filtered_df.copy()
super_df["ingredients"] = filtered_df["ingredients"].apply(parse_ingredients)
```

```python
print("Number of Primary Types :", len(super_df["cuisine"].explode().unique()))
print(super_df["cuisine"].value_counts())
sb.catplot(y = "cuisine", data = super_df, kind = "count", height = 8)
```

```python
def fix_cuisine_list(cuisines):
    if not isinstance(cuisines, list):
        return []
    # Step 1: replace "american" with "north american" because they overlap
    cuisines = ["north-american" if c.lower() == "american" else c for c in cuisines]
    # Step 2: remove duplicates (keep order)
    seen = set()
    new_list = []
    for c in cuisines:
        if c.lower() not in seen:
            new_list.append(c)
            seen.add(c.lower())
    return new_list
df3 = super_df.copy()
df3["cuisine"] = super_df["cuisine"].apply(fix_cuisine_list)
print("Number of Primary Types :", len(df3["cuisine"].explode().unique()))
print(df3["cuisine"].value_counts())
```
